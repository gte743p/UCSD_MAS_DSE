```python
import numpy as np
```

```python
%pylab inline
import matplotlib.pyplot as plt
```

```python
samples_a = np.random.multivariate_normal(mean = [0,0], cov = [[9,0],[0,1]], size = 100)
x_a = [sample[0] for sample in samples_a]
y_a = [sample[1] for sample in samples_a]
```

```python
samples_b = np.random.multivariate_normal(mean = [0,0], cov = [[1,-0.75],[-0.75,1]], size = 100)
x_b = [sample[0] for sample in samples_b]
y_b = [sample[1] for sample in samples_b]
```

```python
plt.plot(x_a, y_a, 'o', alpha = 0.75)
plt.title('Problem 6(a)')
plt.show();
```

```python
plt.plot(x_b, y_b, 'o', alpha = 0.75)
plt.title('Problem 6(b)')
plt.show();
```
