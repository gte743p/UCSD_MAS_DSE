### Plots of mean and std of observations

The five core elements are:

       PRCP = Precipitation (tenths of mm)
       SNOW = Snowfall (mm)
	   SNWD = Snow depth (mm)
       TMAX = Maximum temperature (tenths of degrees C)
       TMIN = Minimum temperature (tenths of degrees C)
       TOBS = Temperature at the time of observation (tenths of degrees C)
       
All are daily measurements.

Jackson, WY weather data obtained from http://www.usclimatedata.com/climate/jackson/wyoming/united-states/uswy0088/2017/1

```python
def plot_mean_std(m,fig,axis):
    if m in ['TMIN','TMAX','TOBS','PRCP']:
        tenths_conv = 0.1
    else:
        tenths_conv = 1.0
    
    mean=STAT[m]['Mean']*tenths_conv
    std=np.sqrt(STAT[m]['Var']*(tenths_conv**2))
    graphs=np.vstack([mean-std,mean,mean+std]).transpose()
    YP.plot(graphs,fig,axis,title='Mean+-std   '+m,labels=['mean-std','mean','mean+std'])
```

```python
plot_pair(['TMIN','TMAX'],plot_mean_std,['Degrees Celsius','Degrees Celsius'])
```

```python
f,a = plt.subplots(figsize=(7,6))
a.set_ylabel('millimeters per day')
plot_mean_std('SNWD',f,a)
```

```python
plot_pair(['TOBS','PRCP'],plot_mean_std,['Degrees Celsius','millimeters per day'])
```

```python
plot_pair(['SNOW', 'SNWD'],plot_mean_std,['millimeters per day','millimeters per day'])
```

### plotting average coeff / measure values over time

```python
meas = 'TOBS'
filename2=data_dir+'/decon_%s_%s.parquet'%(file_index,meas)

df_2=sqlContext.read.parquet(filename2)

#extract longitude and latitude for each station
feature='coeff_3'
sqlContext.registerDataFrameAsTable(df_2,'weather')

Features='station, year, latitude, longitude, elevation, coeff_1, coeff_2, coeff_3, res_1, res_2, res_3'
Query="SELECT %s FROM weather"%Features
pdf = sqlContext.sql(Query).toPandas()
```

```python
year_station_table=pdf.pivot(index='year', columns='station', values='coeff_1')
mean_by_year=np.nanmean(year_station_table,axis=1)
mean_by_station=np.nanmean(year_station_table,axis=0)
```

```python
c1_min_yr = year_station_table.index[np.argwhere(mean_by_year == mean_by_year.min())][0][0]
c1_max_yr = year_station_table.index[np.argwhere(mean_by_year == mean_by_year.max())][0][0]
```

```python
data_dir='../../Data/Weather'
file_index='SBBSSBSB' # my file
#file_index='BBBSBBBB'

#read data
filename=data_dir+'/decon_%s_%s.parquet'%(file_index,meas)

dfp=sqlContext.read.parquet(filename)

# convert to pandas dataframe
dfp = dfp.toPandas()
dfp.columns

dfp = dfp.loc[:,['station','year','latitude','longitude','elevation',
              'coeff_1','coeff_2','coeff_3',
              'res_mean','res_1','res_2','res_3','vector']]

# convert vector to appropriate units
conv = 1.0
if meas in ['TOBS','TMIN','TMAX','PRCP']:
    conv = 0.1
dfp.vector = dfp.vector.apply(lambda row:unpackArray(row,np.float16)*conv)
```

```python
data_dict = {}
for i in range(dfp.shape[0]):
    station = dfp.loc[i,'station']
    year = dfp.loc[i,'year']
    data_dict[(station,year)] = dfp.vector[i]

station_meas = pd.DataFrame.from_dict(data_dict, orient='index')
station = pd.Series(station_meas.index).apply(lambda (s,y):s).values
yrs = pd.Series(station_meas.index).apply(lambda (s,y):y).values
station_meas.loc[:,'station'] = station
station_meas.loc[:,'year'] = yrs
```

```python
meas_by_yr = station_meas.groupby(['year']).mean()
meas_by_yr.loc[:,'mean'] = meas_by_yr.mean(axis=1)
meas_by_yr.loc[:,'min'] = meas_by_yr.min(axis=1)
meas_by_yr.loc[:,'max'] = meas_by_yr.max(axis=1)
```

```python
min_c1_vals = meas_by_yr.loc[int(c1_min_yr),:]
max_c1_vals = meas_by_yr.loc[int(c1_max_yr),:]
plt.plot(min_c1_vals[:365],label=int(c1_min_yr),color='green',alpha=0.7);
plt.plot(max_c1_vals[:365],label=int(c1_max_yr),color='purple',alpha=0.5);
plt.grid();
plt.legend();
plt.title('Daily '+meas+' Value for Years with Min/Max Coeff_1')
plt.ylabel('Degrees Celsius')
plt.xlabel('Day of Year')
plt.show();
```

The first eigenvector for SNWD is positive along its entire range, so positive values for SNWD coefficient 1 will increase SNWD from the mean, and negative values will decrease SNWD from the mean.  From the graph, we see that for any year, the average of the first eigenvector coefficient for SNWD across all stations is positive, with a decreasing trend over time.  This means that we should expect to see a decrease in average SNWD values over time.

This is indeed the case, as we can see from the graph of the average of SNWD readings across all stations by year.  Because the first eigenvector explains roughly 90% of the variance for SNWD, there is a very strong relationship between the first SWND eigenvector coefficient and the direct SNWD readings.

In addition, we should expect to see particularly high and low average SNWD values for the years where the first SNWD eigenvector coefficient is highest and lowest, which are indicated by vertical lines on the graph.  From the graph of daily SNWD readings for 1971 and 1992, we verify that the individual years with the highest and lowest coefficients behave as expected, as the average daily SNWD readings for 1971 are much larger than for 1992.

Furthermore, we can verify this difference statistically by performing a paired t-test with a null hypothesis of equal means:

$$H_0:  \mu_{1934} = \mu_{2008} \\
H_A:  \mu_{1934} \neq \mu_{2008}$$

The paired t-test p-value is $1.52 \times 10^{-83}$, so we can reject $H_0$ in favor of $H_A$.
